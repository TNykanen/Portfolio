```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import datetime
import locale
import requests
import numpy as np
```

```python
#Mielenkiintoiset ID:t: Kotkan satama 530
#Lutakko 938
#Kuopio Jännevirta 827
#Turku Kupittaa 235
#Vaasa 1030
#Suonenjoki 821
#Tampere paasikiventie 438
#Lahti 447
#Oulu Mäntylä 1201
#Rovaniemi Misi 1405

lam_ids = [530, 235, 827, 938]
yearshort = 22
startday = 1
endday = 365

combined_df = pd.DataFrame()

for lam_id in lam_ids:
    for day_number in range(startday, endday+1):
        url = f"https://tie.digitraffic.fi/api/tms/v1/history/raw/lamraw_{lam_id}_{yearshort}_{day_number}.csv"
        column_names = ['pistetunnus', 'vuosi', 'paivan_jarjestysnumero', 'tunti', 'minuutti', 'sekunti', 'sadasosasekunti', 'pituus', 'kaista', 'suunta', 'ajoneuvoluokka', 'nopeus', 'faulty', 'kokonaisaika', 'aikavali', 'jonoalku']
        
        response = requests.get(url)
        if response.status_code == 200:
            df = pd.read_csv(url, delimiter=';', names=column_names)
            combined_df = pd.concat([combined_df, df])
            print(f"Päivän {day_number} data ID:stä {lam_id} haettu ja lisätty yhdistettyyn DataFrameen.")
        else:
            print(f"Virhe haettaessa päivän {day_number} dataa.")

        time.sleep(0.5) #tauko sekunteina

print("Datan hakeminen on valmis.")
```

```python
ajoneuvoluokat = {
    1: 'Henkilö- tai pakettiauto',
    2: 'Kuorma-auto ilman perävaunua',
    3: 'Linja-auto',
    4: 'Kuorma-auto ja puoliperävaunu',
    5: 'Kuorma-auto ja täysperävaunu',
    6: 'Henkilöauto ja peräkärry',
    7: 'Henkilöauto ja asuntovaunu',
    8: 'Moottoripyörät ja mopot',
    9: 'High Capacity Truck'
}

fig, axes = plt.subplots(2, 2, figsize=(12, 12))  # Muutettu koko, jotta tulee kaksi riviä ja kaksi kuvaajaa per rivi

for idx, lam_id in enumerate(lam_ids[:4], 0):  # Käsitellään vain neljä ensimmäistä pistetunnusta
    counts_df = df[df['pistetunnus'] == lam_id].copy()
    counts_df['ajoneuvoluokka'] = counts_df['ajoneuvoluokka'].map(ajoneuvoluokat)
    counts = counts_df['ajoneuvoluokka'].value_counts().sort_index()

    x_ticks = range(len(counts))
    row = idx // 2
    col = idx % 2
    axes[row, col].bar(x_ticks, counts.values)
    axes[row, col].set_ylabel('Ajoneuvojen määrä')
    axes[row, col].set_title('Mittauspiste {}'.format(lam_id))
    axes[row, col].set_xticks(x_ticks)
    axes[row, col].set_xticklabels(counts.index, rotation=90)

    for i, value in enumerate(counts.values):
        axes[row, col].text(i, value, str(value), ha='center', va='bottom')

plt.suptitle('Ajoneuvojen määrä luokittain eri mittauspisteissä {} - {}'.format(date_str1, date_str2), fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()


#ID:t: 
#Kotkan satama 530
#Turku Kupittaa 235
#Vaasa 1030
#Kuopio Jännevirta 827
#Lutakko 938
#Suonenjoki 821
```

```python
lam_ids = df['pistetunnus'].unique()

plt.figure(figsize=(18, 12))

for idx, lam_id in enumerate(lam_ids, 1):
    df_lam_id = df[df['pistetunnus'] == lam_id]
    average_speeds = df_lam_id.groupby('tunti')['nopeus'].mean()
    traffic_counts = df_lam_id.groupby('tunti').size()

    plt.subplot(2, len(lam_ids) // 2, idx)
    
    average_speeds = average_speeds[(average_speeds >= 60) & (average_speeds <= 100)]
    line1, = plt.plot(average_speeds.index, average_speeds.values, marker='o', label='Keskinopeus (km/h)', color='b')
    
    plt.xlabel('Tunti')
    plt.ylabel('Keskinopeus (km/h)')
    plt.title('Mittauspiste {}'.format(lam_id))
    plt.ylim(60, 100)

    ax2 = plt.gca().twinx()
    line2, = ax2.plot(traffic_counts.index, traffic_counts.values, marker='o', label='Liikennemäärä', color='r')
    ax2.set_ylabel('Liikennemäärä')
    ax2.set_ylim(0, max(traffic_counts.values) + 1000)

    plt.legend(handles=[line1, line2], loc='upper right')

plt.suptitle('Liikenteen keskinopeus ja liikennemäärä eri mittauspisteissä', fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

#ID:t: 
#Kotkan satama 530
#Turku Kupittaa 235
#Vaasa 1030
#Kuopio Jännevirta 827
#Lutakko 938
#Suonenjoki 821
```
